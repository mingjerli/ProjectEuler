```python
import urllib.request
f = urllib.request.urlopen('https://projecteuler.net/project/resources/p059_cipher.txt')
data = f.read()
f.close()
```

```python
import binascii
import string

encriptedval = [int(x) for x in data.split(b',')]
```

```python
keys = [ord(c) for c in string.ascii_lowercase]
xormaps = [[x ^ y  for x in encriptedval] for y in keys]
len(xormaps[0])
```

```python
cEw = [ord(c) for c in string.printable]
cEwtest = [[set([y in cEw for y in set([x for x in xormaps[i][j::3]])]) =={True} for i in range(26)] for j in range(3)]
```

```python
ascii0 = [i for i in range(26) if cEwtest[0][i] == True]
ascii1 = [i for i in range(26) if cEwtest[1][i] == True]
ascii2 = [i for i in range(26) if cEwtest[2][i] == True]
indices = [(i0, i1, i2) for i0 in ascii0 for i1 in ascii1 for i2 in ascii2]
ascii0, ascii1, ascii2
```

```python
allpossibletext = [[chr(xormaps[index[i%3]][i]) for i in range(len(xormaps[0]))] for index in indices]
```

```python
allpossibletext = ["".join(allpossibletext[i]) for i in range(len(allpossibletext))]
```
