```python
import numpy as np
import re
import itertools
from itertools import permutations
from collections import Counter, defaultdict
from time import time
```

```python
data = np.genfromtxt('p098_words.txt',delimiter='","',dtype="str").tolist()
data[0] = re.sub('"','',data[0])
data[-1] = re.sub('"','',data[-1])
data[0],data[-1]
```

```python
t0 = time()
N= 10**7
sqnumbers= [str(i*i) for i in range(N)]
squarenumbers = [''.join(sorted(sq)) for sq in sqnumbers]
squarenumberscounter = Counter(squarenumbers)
usefulkeys = [key for key in squarenumberscounter.keys() if squarenumberscounter[key]>1]
sqnumberscounter = [squarenumberscounter[''.join(sorted(sq))] for sq in sqnumbers]
usefulsqnumber = [i*i for i in range(len(sqnumberscounter)) if sqnumberscounter[i]>1]
print("find all possible square numbers at", time()-t0)
del sqnumbers
del squarenumbers
del squarenumberscounter
del sqnumberscounter
```

```python
def isSquare(num):
    from math import sqrt
    sqrtnum = int(sqrt(num))
    return(sqrtnum*sqrtnum ==num)

def findsquarenumber(word):
    result = list()
    wordcounter = Counter(word)
    wordcounterkey = [k for k in wordcounter.keys()]
    ndiffchr = len(wordcounterkey)
    for p in permutations('0123456789', ndiffchr):
        chrdict = dict()
        for i in range(ndiffchr):
            chrdict[wordcounterkey[i]] = p[i]
        num = int(''.join([chrdict[s] for s in word]))
        if len(str(num)) == len(word) and isSquare(num):
#             print(word, num)
            result.append(num)
    return(result)
```

```python
maxval = 0
minnchr = 0
for i, word in enumerate(data):
    if i%10 ==0:
        print(i,word, time()-t0)
    if len(word) >= minnchr:
        sqnumber = findsquarenumber(word)
        if len(sqnumber) > 0:
            for sq in sqnumber:
                if sq > maxval:
                    if sq in usefulsqnumber:
                        maxval = sq
                        minnchr = len(word)
                        print(word, maxval)
```
