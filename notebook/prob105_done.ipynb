```python
import requests
import pandas as pd
from io import StringIO
import itertools
import datetime
```

```python
r = requests.get(url)
```

```python
data = [sorted([int(x) for x in setstr.split(',')]) for setstr in r.text.split('\n')]
```

```python
def isspecialsubset(nums):
    if len(nums)==1:
        return True
    elif len(nums)!=len(set(nums)):
        return False
    else:
        nums = set(nums)
        for i1 in range(1, len(nums)):
            for subset1 in itertools.combinations(nums, i1):
                sum1 = sum(subset1)
                nums1 = set(nums)-set(subset1)
                for i2 in range(i1, len(nums1)+1):
                    for subset2 in itertools.combinations(nums1, i2):
                        sum2 = sum(subset2)
                        if (sum2==sum1) or (i2>i1 and sum1>sum2):
                            print(subset1, subset2)
                            return False
        return True
                    
```

```python
print(isspecialsubset([1]))
print(isspecialsubset([1,2]))
print(isspecialsubset([1,2,3]))
print(isspecialsubset([4,2,3]))
print(isspecialsubset([3,5,6,7]))
print(isspecialsubset([3,4,5,6]))
print(isspecialsubset([6,9,11,12,13]))
print(isspecialsubset([11,17,20,22,23,24]))
print(isspecialsubset([11,18,19,20,22,25]))
```

```python
result = 0
for i, nums in enumerate(data):
    iresult = isspecialsubset(nums)
    if iresult:
        result+=sum(nums)
    print(datetime.datetime.now(), i, result, iresult)
```
