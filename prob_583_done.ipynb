```python
from time import time
from math import gcd, sqrt
```

```python
def isSquare(x):
    testsq = int(sqrt(x))
    return x == testsq*testsq
```

```python
t0 = time()
result=dict()
Lmax = 10 ** 7
count = [0 for _ in range(Lmax+1)]
m_max = int(sqrt(Lmax//2))+1
for m in range(2, m_max):
    for n in range(1, m):
        if gcd(m,n)==1 and (m+n)%2==1:
            LD = m*m+n*n
            L1 = 2*m*n
            L2 = m*m-n*n
            L_tot = LD+L1+L2
            scale = 1
            while L_tot*scale <= Lmax:
                L1s = L1*scale
                L2s = L2*scale
                if str(L1s) not in result.keys():
                    result[str(L1s)] = []
                if str(L2s) not in result.keys():
                    result[str(L2s)] = []
                result[str(L1s)] = result[str(L1s)] + [L2s]
                result[str(L2s)] = result[str(L2s)] + [L1s]
                count[L_tot*scale]+=1
                scale+=1

print(sum([x==1 for x in count]),time()-t0)
```

```python
resultkeys = list(result.keys())
for rkey in resultkeys:
    result[rkey] = sorted(list(set(result[rkey])))
```

```python
x_max = Lmax//4
sumL=0
count=0
t0 = time()

for x in range(1,x_max):
    w_max = 2*x_max-x
    if x%10000 == 0:
        print(x,x_max, count, time()-t0)
#     w_s = getsquaresum(x,w_max)
    if str(x) in result.keys():
        w_s = result[str(x)]
        for w1 in range(len(w_s)-1):
            w = w_s[w1]
            z = int(sqrt(w*w+x*x))
            for w2 in range(w1+1,len(w_s)):
                y = w_s[w2] - w
                if y>= w:
                    if isSquare(y*y+4*x*x):
                        L = (x+y+z)*2
                        if L<=Lmax:
                            sumL+=L
                            count+=1

print(Lmax,sumL, time()-t0, '\n')    
```
