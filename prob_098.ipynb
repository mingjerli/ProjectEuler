```python
import numpy as np
import re
import itertools
from collections import Counter, defaultdict
from time import time
```

```python
data = np.genfromtxt('p098_words.txt',delimiter='","',dtype="str").tolist()
data[0] = re.sub('"','',data[0])
data[-1] = re.sub('"','',data[-1])
data[0],data[-1]
```

```python
len(data)
```

```python
N = 10**7
```

```python
sqnumbers= [i*i for i in range(N)]
squarenumbers = [''.join(sorted(str(i*i))) for i in range(N)]
reversesquarenumbers = squarenumbers[::-1]
squarenumberscounter = Counter(squarenumbers)
usefulkeys = [key for key in squarenumberscounter.keys() if squarenumberscounter[key]>1]
```

```python
len(usefulkeys),len(squarenumbers)
```

```python
dictsquarevalue = defaultdict(int)
count = 1
t0 = time()
for key in usefulkeys:
    if count % 1000 ==0:
        print(count, time()-t0)
    count+=1
    keyindex = len(squarenumbers) - 1 - reversesquarenumbers.index(key)
    keystring = ''.join(map(str,sorted(list(Counter(key).values()))))
    if dictsquarevalue[keystring] < keyindex:
        dictsquarevalue[keystring] = keyindex

# dictsquarevalue.keys()
```

```python
maxindex = 0
maxvalue = 0

for i in range(len(data)):
    keystring = ''.join(map(str,sorted(list(Counter(data[i]).values()))))
    keyvalue = dictsquarevalue[keystring]
    if keyvalue > maxvalue:
        maxindex = i
        maxvalue = keyvalue
        print(i,data[i],keystring, dictsquarevalue[keystring]* dictsquarevalue[keystring], maxvalue*maxvalue)
```
