```python
from math import sqrt
from time import time
```

```python
def squaredigit(x):
    result = 0
    while x>0:
        result += (x%10) * (x%10)
        x//=10
    return(result)
```

```python
N = 10**7
count1 = 0
count89 = 0
t0 = time()
for n in range(1,N+1):
    if n% 100000 == 0:
        print(n, time()-t0,count1, count89)
    sqdigit = squaredigit(n)
    while sqdigit not in set([1,89]):
        sqdigit = squaredigit(sqdigit)
    if sqdigit == 1:
        count1+=1
    if sqdigit == 89:
        count89+=1

print(count1)
print(count89)
```
