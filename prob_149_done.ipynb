```python
# Lagged Fibonacci Generator
# For 1 ≤ k ≤ 55, sk = [100003 − 200003k + 300007k3] (modulo 1000000) − 500000.
# For 56 ≤ k ≤ 4000000, sk = [sk−24 + sk−55 + 1000000] (modulo 1000000) − 500000.
# Thus, s10 = −393027 and s100 = 86613.

def lfg(N=4000000):
    sk = [0 for _ in range(N)]
    for t in range(N):
        k = t+1
        if k<=55:
            sk[t] = (100003 - 200003*k + 300007*k*k*k) % 1000000 - 500000
        else:
            sk[t] = (sk[t-24] + sk[t-55] + 1000000) % 1000000 - 500000
    return(sk)
    
```

```python
def maxsumsubseq(A):
    lenA = len(A)
    S = [0 for _ in range(lenA)]
    S[0] = A[0]
    for i in range(1,lenA):
        if S[i-1] > 0:
            S[i] = S[i-1]+A[i]
        else:
            S[i] = A[i]
    return(max(S))
```

```python
nrow = 2000
M = lfg(nrow*nrow)
```

```python
maxval = M[0]
```

```python
# vertical and hori

for i in range(nrow):
    An = [M[i*(nrow)+j] for j in range(nrow)]
    maxsumsubAn = maxsumsubseq(An)
    if maxsumsubAn > maxval:
        print(maxval,maxsumsubAn,"hori",i)
        maxval = maxsumsubAn
        
    An = [M[i+j*(nrow)] for j in range(nrow)]
    maxsumsubAn = maxsumsubseq(An)
    if maxsumsubAn > maxval:
        print(maxval,maxsumsubAn,"vert",i)
        maxval = maxsumsubAn
```
