```python
# %load peutils/peutils.py
def getPrimes(n):
	import numpy as np
	sieve = np.ones(n/3 + (n%6==2), dtype=np.bool)
	sieve[0] = False
	for i in range(int((n**0.5)/3)+1):
		if sieve[i]:
			k=3*i+1|1
			sieve[int((k*k)/3) :: 2*k] = False
			sieve[int((k*k+4*k-2*k*(i&1))/3) :: 2*k] = False
	return list(np.r_[2,3,((3*np.nonzero(sieve)[0]+1)|1)])

def isPrimes(x):
	from math import sqrt
	testprimes = getPrimes(int(sqrt(x)))
	return all([x % p !=0 for p in testprimes])

# http://stackoverflow.com/questions/34056609/how-to-check-if-an-integer-is-a-perfect-square
def isSquare(x):
	if x == 1:
		return True
	low = 0
	high = x // 2
	root = high
	while root * root != x:
		root = (low + high) // 2
		if low +1 >= high:
			return False
		if root * root > x:
			high = root
		else:
			low = root
	return True

# http://stackoverflow.com/questions/14550794/python-integer-factorization-into-primes
def pFactors(n): 
	"""Finds the prime factors of 'n'""" 
	from math import sqrt 
	pFact, limit, check, num = [], int(sqrt(n)) + 1, 2, n 
	if n == 1: return [1] 
	for check in range(2, limit): 
		while num % check == 0: 
			pFact.append(check) 
			num /= check 
	if num > 1:
		pFact.append(int(num)) 
	return pFact 

# Euler's Totient function, [sometimes called the phi function]
def euler_phi(n):
	from math import gcd
	return(sum([gcd(i,n)==1 for i in range(1,n)]))

if __name__ == "__main__":
	print(getPrimes(100))
	print(isSquare(301))
	print(isSquare(121))
	for i in range(1,1000):
		print (pFactors(i))
```

```python
N = 10 ** 7
midN = math.floor( math.sqrt(N))+1
smallprimelist = getPrimes(midN)[::-1]
primelist = getPrimes(N)
minratio = -1
minn = 1
```
