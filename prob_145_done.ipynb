```python
from time import time
```

```python
def isRevesrible(x):
    return(all([int(s)%2 == 1 for s in str(x)] ))
```

```python
N = 10**9
t0=time()
count=0

for x in range(N):
    if x%10!=0:
        count+=isRevesrible(x+int(str(x)[::-1]))
    if (x+1) % 10**7 ==0:
        print(x+1,count,time()-t0)

print(count)
```
