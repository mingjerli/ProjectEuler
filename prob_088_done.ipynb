```python
from time import time
import itertools
import numpy as np
```

```python
def pFactors(n): 
	"""Finds the prime factors of 'n'""" 
	from math import sqrt 
	pFact, limit, check, num = [], int(sqrt(n)) + 1, 2, n 
	if n == 1: return [1] 
	for check in range(2, limit): 
		while num % check == 0: 
			pFact.append(check) 
			num /= check 
	if num > 1:
		pFact.append(int(num)) 
	return pFact 
```

```python
def pFactors2(n): 
    """Finds the prime factors of 'n'""" 
    from math import sqrt 
    pFact, limit, check, num = [], int(sqrt(n)) + 1, 2, n 
    if n == 1: return [1] 
    for check in range(2, limit): 
        while num % check == 0: 
            pFact.append(check) 
            num /= check 
    if num > 1:
        pFact.append(int(num))
        
    if len(pFact) <= 2:
        return [pFact]
    else:
        result = list()
        for b in set(pFact):
            result.append([b, n//b])
            for l in pFactors(n//b):
                result.append([b]+l)
                for li in range(len(l)):
                    templ = [x for x in l]
                    templ[li] *= b
                    result.append(templ)
        for l in result:
            l.sort()
        result.sort()
        return(list(result for result,_ in itertools.groupby(result)))
```
