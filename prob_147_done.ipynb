```python
from time import time
```

```python
count = 0
t0 = time()
for d1 in range(47):
    for d2 in range(43):
        dim=(d1+1,d2+1)
        for i in range(1,1+dim[0]):
            for j in range(1,1+dim[1]):
                n = (dim[0]-i+1)*(dim[1]-j+1)
                count += n
#             print("hori",i,j, count)

        for up in range(1, 2*i):
            for down in range(1, 2*j):
                dx = up + down
                counta = 0
                for i2 in range(2*i):
                    for j2 in range(2*j):
                        if i2 %2 == j2%2 and i2+dx <= 2*i and j2+up<=2*j and j2-down>=0:
                            counta +=1
                count += counta
#             print("diag",up,down,count)
    print(d1,d2, count,time()-t0)

print(count)
```
