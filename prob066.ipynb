```python
import numpy as np
import math
from time import time
```

```python
def isSquare(x):
    if x == 1:
        return True
    low = 0
    high = x // 2
    root = high
    while root * root != x:
        root = (low + high) // 2
        if low +1 >= high:
            return False
        if root * root > x:
            high = root
        else:
            low = root
    return True

```

```python
maxD = 1000
result = [0 for _ in range(maxD+1)]
testD = [x for x in range(maxD+1)]

# No square sol, assign solution is one .... so it must not be maximum
for i in range(32):
    if i*i < maxD:
        result[i*i] = 1
        testD.remove(i*i)
    
```

```python
x=2
t0 = time()
while len(testD)>1:
    Dy2 = (x+1)*(x-1)
    for D in testD:
        if Dy2 % D ==0:
            if isSquare(Dy2//D):
                result[D] = x
                testD.remove(D)
                print(x, maxD, len(testD), time()-t0)
    x+=1
    if x% 10**7 ==0:
        print(x, maxD, len(testD), time()-t0)

# x=2
# t0 = time()
# while sum([x>0 for x in result[2:]])<maxD-2:
#     Dy2 = (x+1)*(x-1)
#     for D in range(2,maxD+1):
#         if result[D] == 0 and Dy2 % D ==0:
#             if isSquare(Dy2//D):
#                 result[D] = x
#                 print(x, maxD, sum([x>0 for x in result[2:]]), time()-t0)
#     x+=1
#     if x% 10**6 ==0:
#         print(x, maxD, sum([x>0 for x in result[2:]]), time()-t0)
```
